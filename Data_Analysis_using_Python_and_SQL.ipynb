```python
#read data from the file and handle null values
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#rename columns names. make them lower case and replace space with underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
#derive new columns discount, sale price and profit
df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price']= df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
#convert order date from object data type to datetime
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
# Drop columns, ignoring errors if they don't exist
df.drop(columns=['list_price', 'cost_price', 'discount_percent'], inplace=True, errors='ignore')
df.head()
```

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('df_orders')

#load the data into sql server using append option
df.to_sql('df_orders', con=conn , index=False, if_exists = 'replace')
cursor = conn.cursor()

# Execute the SQL query to select all data from the df_orders table
cursor.execute('''SELECT * FROM df_orders
LIMIT 10;''')

# Fetch all rows from the result set
rows = cursor.fetchall()

# Print the results
for row in rows:
    print(row)

# Close the connection
conn.close()
```

```python
# Connect to your SQLite3 database
conn = sqlite3.connect('df_orders')
cursor = conn.cursor()

# Execute the SQL query
cursor.execute('''
WITH cte AS (
    SELECT region, product_id, SUM(sale_price) AS sales
    FROM df_orders
    GROUP BY region, product_id
)
SELECT *
FROM (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY region ORDER BY sales DESC) AS rn
    FROM cte
) A
WHERE rn <= 5;
''')

# Fetch the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
conn.close()
```

```python
# Connect to your SQLite3 database
conn = sqlite3.connect('df_orders')
cursor = conn.cursor()

# Execute the SQL query, using strftime to extract year and month
cursor.execute('''
WITH cte AS (
    SELECT STRFTIME('%Y', order_date) AS order_year,
           STRFTIME('%m', order_date) AS order_month,
           SUM(sale_price) AS sales
    FROM df_orders
    GROUP BY STRFTIME('%Y', order_date), STRFTIME('%m', order_date)
)
SELECT order_month,
       SUM(CASE WHEN order_year = '2022' THEN sales ELSE 0 END) AS sales_2022,
       SUM(CASE WHEN order_year = '2023' THEN sales ELSE 0 END) AS sales_2023
FROM cte
GROUP BY order_month
ORDER BY order_month;
''')

# Fetch the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
conn.close()
```

```python
# Connect to your SQLite3 database
conn = sqlite3.connect('df_orders')
cursor = conn.cursor()

# Execute the SQL query
cursor.execute('''
WITH cte AS (
    SELECT category,
           strftime('%Y%m', order_date) AS order_year_month,
           SUM(sale_price) AS sales
    FROM df_orders
    GROUP BY category, strftime('%Y%m', order_date)
)
SELECT *
FROM (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY sales DESC) AS rn
    FROM cte
) A
WHERE rn = 1;
''')

# Fetch the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
conn.close()
```

```python
# Connect to your SQLite3 database
conn = sqlite3.connect('df_orders')
cursor = conn.cursor()

# Execute the SQL query
cursor.execute('''
WITH cte AS (
    SELECT sub_category, STRFTIME('%Y', order_date) AS order_year, SUM(sale_price) AS sales
    FROM df_orders
    GROUP BY sub_category, STRFTIME('%Y', order_date)
),
cte2 AS (
    SELECT sub_category,
           SUM(CASE WHEN order_year = '2022' THEN sales ELSE 0 END) AS sales_2022,
           SUM(CASE WHEN order_year = '2023' THEN sales ELSE 0 END) AS sales_2023
    FROM cte
    GROUP BY sub_category
)
SELECT sub_category, (sales_2023 - sales_2022) AS growth
FROM cte2
ORDER BY growth DESC
LIMIT 1;
''')

# Fetch the result
result = cursor.fetchone()

# Print the result
print(result)

# Close the connection
conn.close()
```
